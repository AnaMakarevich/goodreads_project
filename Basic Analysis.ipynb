```python
q2 = """
SELECT bc.club_name, bc.book_club_id, b.pub_year
FROM genre g
JOIN genre_book gb
ON gb.genre_id=g.genre_id
JOIN book b
ON b.book_id=gb.book_id
JOIN book_club_book bcb
ON bcb.book_id=b.book_id
JOIN book_club bc
ON bc.book_club_id=bcb.book_club_id
"""
#extr = GROUP BY b.pub_year, bc.club_name
```

```python
df = get_query_results(q2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import plotly.graph_objs as go
import plotly.plotly as py
from plotly import tools
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
df = df[df.pub_year > 1950]
```

```python
club_names = df.club_name.unique()
club_names
```

```python
data = []
for i in range(4):
    trace = go.Box(
        y=df[df.club_name==club_names[i]].pub_year,
        name = club_names[i]
    )
    data.append(trace)
#data = [trace0, trace1, trace2, trace3]
layout = go.Layout(
)
fig = go.Figure(data=data, layout=layout)
iplot(fig,config={'showLink': False})
```

```python
hist_data = [df[df.club_name==club_names[i]].pub_year for i in range(4)]
bins_setup = dict(
        start=1950,
        end=2020,
        size=2
    )

hist_figures = [
    go.Histogram(
    x=hist_data[i],
    opacity=0.75,
    name = club_names[i],
    xbins = bins_setup) for i in range(4)]

layout = go.Layout(barmode='overlay')
fig = go.Figure(data=hist_figures, layout=layout)

iplot(fig, filename='overlaid histogram',config={'showLink': False})
```
